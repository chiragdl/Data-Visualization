```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

Question 1:



plot a line plot between a and b:


a=np.arange(40,50)


b=np.arange(50,60)

```python
a=np.arange(40,50)
b=np.arange(50,60)

plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='Red')
plt.plot(a,b,'g*-',linewidth=1,markersize=8)
plt.show()
```

Question 2:
Plot a line plot showing the sales trend in company 1 and 2:

days = [1,2,3,4,5,6,7] #days of d week
sales_1 = [160,150,140,145,175,165,180] #sales of company1
sales_2 = [70,90,160,150,140,145,175] #sales of company2

```python
days = [1,2,3,4,5,6,7]
sales_1 = [160,150,140,145,175,165,180] #company1
sales_2 = [70,90,160,150,140,145,175] #company2

plt.xlabel("X-Axis", color='red')
plt.ylabel("Y-Axis", color='red')

plt.plot(days, sales_1, 'ro-', markersize=12, linewidth=2)
plt.plot(days, sales_2, 'go-', markersize=12, linewidth=2)
plt.show()
```
