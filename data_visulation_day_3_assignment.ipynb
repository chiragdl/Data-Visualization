To Plot using sns.relplot, line plot, box plot on fmri data set

```python
import numpy as np
import matplotlib as mpl
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from numpy.random import randn,randint,uniform,sample
%matplotlib inline
```

```python
fmri= sns.load_dataset('fmri')
fmri.head()
```

```python
sns.relplot(y='signal',x='timepoint',kind='line',hue='event',size='region',data=fmri)
```

```python
sns.relplot(y='signal',x='timepoint',kind='line',hue='subject',size='region',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',hue='subject',style='event',size='region',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',hue='event',style='subject',size='region',data=fmri)
```

```python
sns.lineplot(y='signal',x='timepoint',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',hue='subject',style='event',size='region',data=fmri,kind='line')
```

```python
sns.relplot(x='timepoint',y='signal',hue='event',style='subject',size='region',data=fmri,kind='line')
```
